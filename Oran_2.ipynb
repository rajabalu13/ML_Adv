DISCRIMINANT ANALYSIS

```python
import numpy as np
import pandas as pd

data = pd.read_csv('Ricenwheat.csv')
data
```

```python
data = data.replace([' ''.''-'] , np.nan)
replace = data['Production'].mean()
data.fillna(replace , inplace = True)

data = data.replace([' ''.''-'] , np.nan)
replace = data['Area'].mean()
data.fillna(replace , inplace = True)

data
```

```python
#data preprocessing
from sklearn.preprocessing import LabelEncoder

var_mod = ['State_Name','District_Name','Season','Crop']
le = LabelEncoder()
for i in var_mod:
    data[i] = le.fit_transform(data[i])
```

```python
a = data.iloc[:,0:6].values
b = data.iloc[:,4].values
```

```python
a.shape
```

```python
b.shape
```

```python
a.shape[0] == b.shape[0]
```

```python
#spilts dataset into training and test sets
from sklearn.model_selection import train_test_split

a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=0)
```

```python
#performing feature scaling
from sklearn.preprocessing import StandardScaler 

sc = StandardScaler()
a_train = sc.fit_transform(a_train)
a_test = sc.transform(a_test)
#print(X_train)
```

```python
#Performing LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as lda

lda = lda(n_components=1)
a_train = lda.fit_transform(a_train, b_train)
a_test = lda.transform(a_test)
```

```python
#Training the dataset
#Fitting thr model 
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)

classifier.fit(a_train, b_train)
b_pred = classifier.predict(a_test)
```

```python
#Evaluating the performance with confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(b_test, b_pred)
print(cm)
#finds the accuracy of the matrix
print('Accuracy' + str(accuracy_score(b_test, b_pred)))
```

#LOGISTIC REGRESSION

```python
X = data[['State_Name', 'District_Name' , 'Crop_Year', 'Season','Area', 'Production']]
Y = data['Crop']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
from sklearn.metrics import accuracy_score,roc_curve,roc_auc_score
from sklearn.linear_model import LogisticRegression 

#Fitting the model
model = LogisticRegression()
model.fit(x_train,y_train)
predictions = model.predict(x_test)
print(accuracy_score(y_test, predictions))
```

```python
import matplotlib.pyplot as plt

#ROC Curve
pred_lr = model.predict_proba(x_test)[:,1]
fpr_lr,tpr_lr,_ = roc_curve(y_test,pred_lr)

fig, ax = plt.subplots()
ax.plot(fpr_lr,tpr_lr)
plt.plot(fpr_lr, tpr_lr, marker='.', label='Logistic')
lr_auc = roc_auc_score(y_test, pred_lr)
print('Logistic: ROC AUC=%.3f' % (lr_auc))
plt.legend()
plt.show()
```
